Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений



```python
import pandas as pd
```

```python
df_log = pd.read_csv('visit_log.csv', sep = ';')
df_log.head(10)
```

```python
df_log.loc[(df_log.traffic_source == 'yandex') & (df_log.traffic_source == 'google'), 'source_type'] = 'organic'
df_log.loc[(df_log.region == 'Russia') & (df_log.traffic_source == 'paid'), 'source_type'] = 'ad'
df_log.loc[(df_log.region == 'Russia') & (df_log.traffic_source == 'email'), 'source_type'] = 'ad'
df_log.loc[(df_log.region != 'Russia') & (df_log.traffic_source == 'paid'), 'source_type'] = 'other'
df_log.loc[(df_log.region != 'Russia') & (df_log.traffic_source == 'email'), 'source_type'] = 'other'
df_log.loc[(df_log.traffic_source != 'yandex') & (df_log.traffic_source != 'google') 
           & (df_log.traffic_source != 'paid') & (df_log.traffic_source != 'email'),'source_type'] = df_log.traffic_source
df_log.head(20)
```

```python
def source_type(row):
    if 'Russia' in row['region']:
        if 'paid' or 'email' in row['traffic_source']:
            return 'ad'
        elif 'yandex' or 'google' in row['traffic_source']:
            return 'organic'
    elif 'paid' in row['traffic_source']:
        return 'other'
    else:
        return row['traffic_source']
```

```python
df_log['source_type'] = df_log.apply(source_type, axis=1)
df_log.head(20)
#ЗАДАЧА 1 решена
```

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings
```

```python
df_ratings_grouped = df_ratings.groupby('userId').agg(['count', 'min', 'max'])['timestamp'].reset_index()
df_ratings_grouped
```

```python
df_ratings_100 = df_ratings_grouped[df_ratings_grouped['count'] > 100]
df_ratings_100.rename(columns={'count': 'rating_count', 'min': 'first_visit_date', 'max': 'last_visit_date'}, inplace=True)
df_ratings_100
```

```python
df_ratings_100['ave_lifetime'] = df_ratings_100['last_visit_date'] - df_ratings_100['first_visit_date']
ave_lifetime_of_all = df_ratings_100['ave_lifetime'].mean()
print('Среднее время жизни активных пользователей -', ave_lifetime_of_all)
# Почему-то среднее время жизни считается то в формате юникстайм - если перезапустить ядро и выполнить код, 
#    то в формате даты 463 days 21:28:27.449612384 - если в формулу выше поставить mean() / 3600 / 24. Непонятно.
# Переводим столбцы с датами входа, выхода в формат обычной даты - для удобства восприятия в табличном виде
df_ratings_100['first_visit_date'] = pd.to_datetime(df_ratings_100['first_visit_date'], unit='s')
df_ratings_100['last_visit_date'] = pd.to_datetime(df_ratings_100['last_visit_date'], unit='s')
df_ratings_100
```

аналогичную таблицу по типам выручки с указанием адреса клиента

Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки
auto - автомобильные перевозки
air - воздушные перевозки
client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
joined_3 = joined_2.merge(client_base, how='left', on='client_id')
joined_3[['client_id', 'address', 'rzd_revenue', 'auto_revenue', 'air_revenue']]
#ЗАДАЧА 4 решена
```
